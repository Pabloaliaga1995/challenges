```python
import pandas as pd
```

```python
df1 = pd.read_csv('../pandas-apply-lambda-master/data/input/IMDB-Movie-Data.csv')
df1
```

```python
def func_bins(vote):
    vote = int(vote)
    if vote < 999:
        return 1
    elif vote >= 1000 and vote < 9999:
        return 2
    elif vote >=10000 and vote < 99999:
        return 3
    elif vote >=100000 and vote < 999999:
        return 4
    else:
        return 5
    
```

```python
df1['bins'] = df1['Votes'].apply(func_bins)
```
